```python
import numpy as np
import matplotlib.pyplot as plt

import h5py

import pycwt
```

```python
noise_dataset = h5py.File(r'F:\PythonProjects\CCLSN\2-CWT\inputdata\YS_noise.h5', "r")
event_dataset = h5py.File(r'F:\PythonProjects\CCLSN\2-CWT\inputdata\YS_event.h5', "r")

noise_data = np.array(noise_dataset["noise"][:])
event_data = np.array(event_dataset["event"][:])
```

```python
event_data.shape
```

```python
noise_data.shape
```

```python
event = event_data[1,:,0]
[cwt, frequencies] = pycwt.cwt4series(event,sampling_rate = 100)
cwtmatr = pycwt.subs_matr(cwt,m=20)
```

```python
t = np.arange(0, 40, 1.0 / 100)

t1 = np.arange(0, 20, 1.0 / 10)
```

```python
t.shape
```

```python
t1.shape
```

```python
plt.figure(figsize=(8, 8))
plt.subplot(211)
plt.plot(t, event)

plt.subplot(212)
plt.contourf(t1, frequencies, abs(cwtmatr))
#cb = plt.colorbar()
plt.show()
```

```python
noise = noise_data[3,:,0]
[cwt, frequencies] = pycwt.cwt4series(noise,sampling_rate = 100)
cwtmatr = pycwt.subs_matr(cwt)
```

```python
plt.figure(figsize=(8, 16))
plt.subplot(211)
plt.plot(t, noise)

plt.subplot(212)
plt.contourf(t1, frequencies, abs(cwtmatr))
plt.show()
```
